```python
import numpy as np
import os
import cv2
from matplotlib import pyplot as plt
from tensorflow.keras.models import load_model

# Load your pre-trained model (replace with your model file path)
model = load_model('C:/Users/Hp/Desktop/review/facial/model.h5')  # Ensure this is the correct path to your model

# Define a mapper for class indices to emotion labels
mapper = {
    0: 'Angry',
    1: 'Disgust',
    2: 'Fear',
    3: 'Happy',
    4: 'Sad',
    5: 'Surprise',
    6: 'Neutral'
}

# Directory containing test images
test_images_dir = 'C:/Users/Hp/Desktop/review/facial/image'
  # Use forward slashes for compatibility

# Check if the directory exists
if not os.path.exists(test_images_dir):
    print("Directory does not exist!")
else:
    # Load all images in the test directory
    image_files = [f for f in os.listdir(test_images_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]

    # Check if there are images to process
    if len(image_files) == 0:
        print("No image files found in the directory.")
    else:
        # Set up the figure for displaying predictions
        num_images = len(image_files)
        fig, axs = plt.subplots(2, num_images, figsize=(15, 5))  # Adjust the figure size as needed

        # Iterate through the images and make predictions
        for i, image_file in enumerate(image_files):
            image_path = os.path.join(test_images_dir, image_file)

            # Load and preprocess the image
            sample_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Use the dynamic image_path
            if sample_img is None:
                print(f"Error loading image: {image_path}")
                continue  # Skip to the next image if there’s an error
            
            sample_img = cv2.resize(sample_img, (48, 48))  # Resize to match model input
            sample_img = sample_img / 255.0  # Normalize the image
            sample_img = np.expand_dims(sample_img, axis=-1)  # Add channel dimension
            sample_img = np.expand_dims(sample_img, axis=0)  # Add batch dimension

            # Get prediction
            prediction = np.argmax(model.predict(sample_img), axis=1)[0]
            predicted_emotion = mapper.get(prediction, 'Unknown')  # Get the emotion label
            print(f"Image '{image_file}' predicted as class: {prediction} ({predicted_emotion})")

            # Display the image and prediction result
            axs[0, i].imshow(sample_img[0, :, :, 0], cmap='gray')
            axs[0, i].set_xticks([])
            axs[0, i].set_yticks([])
            axs[0, i].set_title(f"Pred: {predicted_emotion}")  # Show the emotion label

        plt.tight_layout()
        plt.show()
```
